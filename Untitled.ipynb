```python
import pandas as pd
import numpy as np
import seaborn as sns

df=pd.read_csv("food-price-index-2021.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Data_value"]=df["Data_value"].fillna(df["Data_value"].median())
```

```python
df.isnull().sum()
```

```python
df.boxplot()
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df_out = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df_out.shape)
```
